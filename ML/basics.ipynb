```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
age = [25, 26, 27, 28, 29, 30]
salary = [45000, 53000, 65000, 75000, 85000, 95000]
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(age, salary)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

## We can change the color of dots by

```python
plt.scatter(age, salary, c='yellow')
```

```python
plt.scatter(age, salary, s=50)
plt.grid()
```

```python
plt.figure(figsize=(10, 8))
plt.plot(age, salary)
```
