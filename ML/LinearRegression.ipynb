```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('N:/PycharmProjects/AI-ML_Internship/assets/cost_revenue_clean.csv')
df.head()
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(df['production_budget_usd'], df['worldwide_gross_usd'])
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```

```python
X = df.drop('worldwide_gross_usd', axis='columns')
Y = df['worldwide_gross_usd']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression(n_jobs=-1)
```

```python
regressor.fit(X, Y)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(X, Y)
plt.plot(X ,regressor.predict(X), c='red')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```
